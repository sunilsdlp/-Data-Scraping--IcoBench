```python
'''
Author:Sunil Kumar
This Code scraps the Details(ICO Name,Start_Date,End_Date,Rating) from IcoBench.com
'''
import requests
import pandas as pd
from bs4 import BeautifulSoup
page=requests.get('https://icobench.com/icos?page=1')   #Link of the page
soup=BeautifulSoup(page.content,'lxml')
table_1=soup.find('table')
table_row=table_1.find_all('tr')
ico_list=[]
date_list=[]
for tr in table_row:
    for td in list(tr.find_all('td')):
        for a in list(td.find_all('a',class_='name')):
            ico_list.append(a.text.strip())
        for date in list(td.find_all('div',class_='row')):
            date_list.append(date.text.strip()) 
st = date_list[0]
start, end, rate = [], [], []
for i in range(len(date_list)):
    d = date_list[i].split(": ")
    if d[0] == "Start":
        start.append(d[-1])
    if d[0] == "End":
        end.append(d[-1])
    if d[0] == "Rate":
        rate.append(d[-1])
df = pd.DataFrame()
df["Name"], df["start"], df["end"], df["rate"] =ico_list, start, end, rate
```
